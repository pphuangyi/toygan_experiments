```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
%matplotlib inline
```

```python
path_base = Path('/home/yhuang2/PROJs/LS4GAN/toygan_outdir/toy-adc/bc/transfer/bc-cyclegan-transfer-8x8/')

model_prefix = 'model_d(toyzero-precropped)_md(None)_mg(unet_128)'

losses_cyclegan = pd.read_csv(path_base/f'{model_prefix}_reinitialized/history.csv')
losses_pretrain_full = pd.read_csv(path_base/f'{model_prefix}_zero/history.csv')
losses_pretrain_epoch5 = pd.read_csv(path_base/f'{model_prefix}_pretrain-epoch-5/history.csv')
losses_pretrain_wgen = pd.read_csv(path_base/f'{model_prefix}_wgen/history.csv')

dfs = [
    # losses_cyclegan,
    # losses_pretrain_full,
    losses_pretrain_wgen,
]

labels = [
    # 'from scratch', 
    # 'pretrain_lsgan', 
    'pretrain_wgen'
]
```

```python
# combine generator loss:
def overall_gen_loss(df, l, p=.5):
    # l = lambda, p = phi
    df['overall_gen_loss'] = df['gen_ab'] + df['gen_ba'] + l * (df['cycle_a'] + df['cycle_b'])
    if 'idt_a' in df.columns:
        df['overall_gen_loss'] += p * l * (df['idt_a'] + df['idt_b'])
        
l, p = 10, .5
for df in dfs:
    overall_gen_loss(df, l, p)
    
for i in range(len(dfs)):
    df = dfs[i]
    dfs[i] = df[df['epoch'] > 25]
```

```python
def plot_losses(dfs, labels, loss_type, two_domains, fname):
    if two_domains:
        fig, axes = plt.subplots(1, 2, figsize=(10, 4))

    ax = axes[0]
    for df, label in zip(dfs,labels):
        y_col = f'{loss_type}_a' if loss_type != 'gen' else f'{loss_type}_ab'
        df.plot(x='epoch', y=y_col, ax=ax, label=label)
        ax.set_ylabel(f'{loss_type} losses on fake')
        
    ax = axes[1]
    for df, label in zip(dfs,labels):
        y_col = f'{loss_type}_a' if loss_type != 'gen' else f'{loss_type}_ba'
        df.plot(x='epoch', y=y_col, ax=ax, label=label)
        ax.set_ylabel(f'{loss_type} losses on real')
    
    fig.suptitle(f'CyclgeGAN {loss_type} loss comparison', fontsize=15)
    # fig.savefig(fname, bbox_inches='tight', dpi=300)
    
plot_losses(dfs, labels, 'cycle', two_domains=True, fname=None)
plot_losses(dfs, labels, 'idt', two_domains=True, fname=None)

plot_losses(dfs, labels, 'disc', two_domains=True, fname=None)
plot_losses(dfs, labels, 'gen', two_domains=True, fname=None)
```

```python
loss_type = 'cycle'

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

ax = axes[0]
losses_cyclegan.plot(x='epoch', y=f'{loss_type}_a', ax=ax, label='from scratch')
losses_pretrain_full.plot(x='epoch', y=f'{loss_type}_a', ax=ax, label='with pretraining')
losses_pretrain_epoch5.plot(x='epoch', y=f'{loss_type}_a', ax=ax, label='with pretraining epoch 5')
ax.set_ylabel(f'{loss_type} losses on fake')

ax = axes[1]
losses_cyclegan.plot(x='epoch', y=f'{loss_type}_b', ax=ax, label='from scratch')
losses_pretrain_full.plot(x='epoch', y=f'{loss_type}_b', ax=ax, label='with pretraining')
losses_pretrain_epoch5.plot(x='epoch', y=f'{loss_type}_b', ax=ax, label='with pretraining epoch 5')
ax.set_ylabel(f'{loss_type} losses on real')

fig.suptitle(f'CyclgeGAN {loss_type} loss comparison', fontsize=15)
fig.savefig(f'results/cyclegan_loss_comparison_{loss_type}.png', bbox_inches='tight', dpi=300)
```
