```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
losses_from_scratch = pd.read_csv('../cyclegan/from_scratch/model_d(toyzero-preunaligned)_md(None)_mg(unet_128)_fdeb0c30c293f68c82125afff84e6717/history.csv')
losses_with_pretraining_8x8 = pd.read_csv('../cyclegan/with_pretraining/model_d(toyzero-preunaligned)_md(None)_mg(unet_128)_7ca272aabcc013997952c05b2a0ce47d/history.csv')
losses_with_pretraining_1x1 = pd.read_csv('../cyclegan/with_pretraining/bc-cyclegan-50k-1x1/model_d(toyzero-preunaligned)_md(None)_mg(unet_128)_e63d43e277f6ca2ce121ad435cacf167/history.csv')
```

```python
losses_from_scratch = losses_from_scratch[losses_from_scratch['epoch'] > 25]
losses_with_pretraining_8x8 = losses_with_pretraining_8x8[losses_with_pretraining_8x8['epoch'] > 25]
losses_with_pretraining_1x1 = losses_with_pretraining_1x1[losses_with_pretraining_1x1['epoch'] > 25]
```

```python
losses_from_scratch.head()
```

```python
loss_type = 'cycle'

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

ax = axes[0]
losses_from_scratch.plot(x='epoch', y=f'{loss_type}_a', ax=ax, label='from scratch')
losses_with_pretraining_8x8.plot(x='epoch', y=f'{loss_type}_a', ax=ax, label='with pretraining 8x8')
losses_with_pretraining_1x1.plot(x='epoch', y=f'{loss_type}_a', ax=ax, label='with pretraining 1x1')
ax.set_ylabel(f'{loss_type} losses on fake')

ax = axes[1]
losses_from_scratch.plot(x='epoch', y=f'{loss_type}_b', ax=ax, label='from scratch')
losses_with_pretraining_8x8.plot(x='epoch', y=f'{loss_type}_b', ax=ax, label='with pretraining 8x8')
losses_with_pretraining_1x1.plot(x='epoch', y=f'{loss_type}_b', ax=ax, label='with pretraining 1x1')
ax.set_ylabel(f'{loss_type} losses on real')

fig.suptitle(f'CyclgeGAN {loss_type} loss comparison', fontsize=15)
fig.savefig(f'results/cyclegan_loss_comparison_{loss_type}.png', bbox_inches='tight', dpi=300)
```
